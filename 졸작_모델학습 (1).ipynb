### 2-3 data를 독립변수(x)와 종속변수(y)로 나누기

```python
X = data.drop(['Y'], axis=1)
Y = data['Y']
```

### 2-4 train, test data로 데이터셋 구분

```python
# 학습에 사용할 X-train, Y-train data와 평가에 사용할 X-test, Y-test로 데이터셋을 나누기
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=2/10,
                                                    stratify=Y,
                                                    random_state=10)
#test_size=3/10, 전체 데이터셋에서 test 데이터를 30% 사용하겟다. 나머지 70%는 train set
#stratify : train, test를 나눌 때 Y 클래스 비율을 각각 동일하게 반영하겠다는 뜻 
#random_state : 이걸 특정 값으로 지정해 놓지 않으면 매번 모델을 만들때마다 랜덤하게 데이터를 뽑으면서
                #모델이 매번 달라짐. 따라서 하이퍼파라미터 튜닝시에 값을 고정하고 튜닝해야함
```

### 2-4 모델 학습

```python
# 머신러닝 모델 불러오기. 로지스틱, 의사결정나무, 랜덤포레스트, 그래디언트 부스팅 4개 모델 사용
LR = LogisticRegression(random_state = 0)
DT = DecisionTreeClassifier(random_state = 0, criterion = 'entropy', max_depth= 4)
RF = RandomForestClassifier(random_state = 0, criterion = 'gini', max_depth = None, n_estimators= 59, max_features='sqrt')
GB = GradientBoostingClassifier(random_state = 0)

# 머신러닝 모델 train_data로 학습하기
LR.fit(X_train, Y_train)
DT.fit(X_train, Y_train)
RF.fit(X_train, Y_train)
GB.fit(X_train, Y_train)
```

```python
DT_best = DecisionTreeClassifier(random_state=0)

param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [None, 1, 2, 3, 4, 5, 6, 7, 8, 9]}

DT_cv = GridSearchCV(estimator=DT_best, param_grid=param_grid, cv=5)

DT_cv.fit(X_train, Y_train)
```

```python
print('Best parameters (Grid Search): \n',
      DT_cv.best_params_)
```

```python
# 최적 하이퍼파라미터로 학습한 모델. lg_cv.best_estimator_.predict() 통해 사용 가능
DT_cv.best_estimator_
```

```python
# 성능 평가하기. predict 함수가 모델에 인풋 넣었을 떄 Y 뱉어내는 것. 즉 모델의 아웃풋임

# 인풋으로 X_train 넣었을 때
LR_y_train_pred = LR.predict(X_train)
DT_y_train_pred = DT.predict(X_train)
RF_y_train_pred = RF.predict(X_train)
GB_y_train_pred = GB.predict(X_train)

# 인풋으로 X_test 넣었을 때. 평가시에는 이 값을 활용해야함
LR_y_test_pred = LR.predict(X_test)
DT_y_test_pred = DT.predict(X_test)
RF_y_test_pred = RF.predict(X_test)
GB_y_test_pred = GB.predict(X_test)
```

```python
# 성능평가 결과
# 분류문제 성능평가시 사용하는 Accuracy, Precision, Recall, F1-score, Balanced_score 5개 사용
print("train_data")
print('- LR_Accuracy (Train) : {:.4}'.format(accuracy_score(Y_train, LR_y_train_pred)))
print('- LR_Precision score (Train) : {:.4}'.format(precision_score(Y_train, LR_y_train_pred)))
print('- LR_recall score (Train) : {:.4}'.format(recall_score(Y_train, LR_y_train_pred)))
print('- LR_F1 score (Train) : {:.4}'.format(f1_score(Y_train, LR_y_train_pred)))
print('- LR_Balanced score (Train) : {:.4}'.format(balanced_accuracy_score(Y_train, LR_y_train_pred)))
print("###############")
print('- DT_Accuracy (Train) : {:.4}'.format(accuracy_score(Y_train, LR_y_train_pred)))
print('- DT_Precision score (Train) : {:.4}'.format(precision_score(Y_train, DT_y_train_pred)))
print('- DT_recall score (Train) : {:.4}'.format(recall_score(Y_train, DT_y_train_pred)))
print('- DT_F1 score (Train) : {:.4}'.format(f1_score(Y_train, DT_y_train_pred)))
print('- DT_Balanced score (Train) : {:.4}'.format(balanced_accuracy_score(Y_train, DT_y_train_pred)))
print("###############")
print('- RF_Accuracy (Train) : {:.4}'.format(accuracy_score(Y_train, RF_y_train_pred)))
print('- RF_Precision score (Train) : {:.4}'.format(precision_score(Y_train, RF_y_train_pred)))
print('- RF_recall score (Train) : {:.4}'.format(recall_score(Y_train, RF_y_train_pred)))
print('- RF_F1 score (Train) : {:.4}'.format(f1_score(Y_train, RF_y_train_pred)))
print('- RF_Balanced score (Train) : {:.4}'.format(balanced_accuracy_score(Y_train, RF_y_train_pred)))
print("###############")
print('- GB_Accuracy (Train) : {:.4}'.format(accuracy_score(Y_train, GB_y_train_pred)))
print('- GB_Precision score (Train) : {:.4}'.format(precision_score(Y_train, GB_y_train_pred)))
print('- GB_recall score (Train) : {:.4}'.format(recall_score(Y_train, GB_y_train_pred)))
print('- GB_F1 score (Train) : {:.4}'.format(f1_score(Y_train, GB_y_train_pred)))
print('- GB_Balanced score (Train) : {:.4}'.format(balanced_accuracy_score(Y_train, GB_y_train_pred)))

print("###############")
print("test_data")
print("###############")
print('- LR_Accuracy (Test) : {:.4}'.format(accuracy_score(Y_test, LR_y_test_pred)))
print('- LR_Precision score (Test) : {:.4}'.format(precision_score(Y_test, LR_y_test_pred)))
print('- LR_recall score (Test) : {:.4}'.format(recall_score(Y_test, LR_y_test_pred)))
print('- LR_F1 score (Test) : {:.4}'.format(f1_score(Y_test, LR_y_test_pred)))
print('- LR_Balanced score (Test) : {:.4}'.format(balanced_accuracy_score(Y_test, LR_y_test_pred)))
print("###############")
print('- DT_Accuracy (Test) : {:.4}'.format(accuracy_score(Y_test, DT_y_test_pred)))
print('- DT_Precision score (Test) : {:.4}'.format(precision_score(Y_test, DT_y_test_pred)))
print('- DT_recall score (Test) : {:.4}'.format(recall_score(Y_test, DT_y_test_pred)))
print('- DT_F1 score (Test) : {:.4}'.format(f1_score(Y_test, DT_y_test_pred)))
print('- DT_Balanced score (Test) : {:.4}'.format(balanced_accuracy_score(Y_test, DT_y_test_pred)))
print("###############")
print('- RF_Accuracy (Test) : {:.4}'.format(accuracy_score(Y_test, RF_y_test_pred)))
print('- RF_Precision score (Test) : {:.4}'.format(precision_score(Y_test, RF_y_test_pred)))
print('- RF_recall score (Test) : {:.4}'.format(recall_score(Y_test, RF_y_test_pred)))
print('- RF_F1 score (Test) : {:.4}'.format(f1_score(Y_test, RF_y_test_pred)))
print('- RF_Balanced score (Test) : {:.4}'.format(balanced_accuracy_score(Y_test, RF_y_test_pred)))
print("###############")
print('- GB_Accuracy (Test) : {:.4}'.format(accuracy_score(Y_test, RF_y_test_pred)))
print('- GB_Precision score (Test) : {:.4}'.format(precision_score(Y_test, RF_y_test_pred)))
print('- GB_recall score (Test) : {:.4}'.format(recall_score(Y_test, RF_y_test_pred)))
print('- GB_F1 score (Test) : {:.4}'.format(f1_score(Y_test, RF_y_test_pred)))
print('- GB_Balanced score (Test) : {:.4}'.format(balanced_accuracy_score(Y_test, RF_y_test_pred)))
print("###############")
```
